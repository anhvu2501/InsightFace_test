```python
from deepface import DeepFace
```

```python
img1_path = "/kaggle/input/test-face-reg/img1.jpg"
img2_path = "/kaggle/input/test-face-reg/img2.jpg"
```

```python
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread(img1_path)
img2 = cv2.imread(img2_path)
```

```python
plt.imshow(img1[:, :, ::-1])
plt.show()
```

```python
plt.imshow(img2[:, :, ::-1])
plt.show()
```

```python
# Face verification
result = DeepFace.verify(img1_path, img2_path, model_name='ArcFace', detector_backend='retinaface')
print(result)
print(result['verified'])
```

```python
def face_verify(img1_path, img2_path):
    img1 = cv2.imread(img1_path)
    img2 = cv2.imread(img2_path)
    
    plt.imshow(img1[:, :, ::-1])
    plt.show()
    plt.imshow(img2[:, :, ::-1])
    plt.show()
    
    result = DeepFace.verify(img1_path, img2_path, model_name='ArcFace', detector_backend='retinaface')
    print("Face verification result: ", result)
    
    verification = result['verified']
    
    if verification:
        print("They are the same")
    else:
        print("They are not the same")
```

```python
face_verify(img1_path, img2_path)
```
